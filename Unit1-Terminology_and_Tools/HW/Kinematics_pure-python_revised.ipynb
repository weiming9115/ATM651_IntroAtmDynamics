### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= \frac{\partial (U+c_{1}x+c_{2}y)}{\partial x} + \frac{\partial (V+c_{3}x+c_{4}y)}{\partial y} = c_{1} + c_{4} $
* $ \zeta_{rel} = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} = \frac{\partial (V+c_{3}x+c_{4}y)}{\partial x} - \frac{\partial (U+c_{1}x+c_{2}y)}{\partial y} = c_{2}-c_{3}$
* $ def1 = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y} = \frac{\partial (U+c_{1}x+c_{2}y)}{\partial x} - \frac{\partial (V+c_{3}x+c_{4}y)}{\partial y} = c_{1}-c_{4}$
* $ def2 = \frac{\partial v}{\partial x} + \frac{\partial u}{\partial y} = \frac{\partial (V+c_{3}x+c_{4}y)}{\partial x} + \frac{\partial (U+c_{1}x+c_{2}y)}{\partial y} = c_{2}+c_{3}$

## Exercise 2: Create and display the basic fields
* First, execute a code cell setting up a coordinate system (x,y) from -10 to 10
* Next, execute the cell defining the basic "pure" div, vort, def1, def2, and U,V fields 
* Finally, display them. The first one has been done for you. You complete the work. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

```python
# Define the display field udisp,vdisp and give it a label. 
# Edit-> Copy and paste this cell, then adjust the top 2 lines for the other cases

udisp_a,vdisp_a = udiv, vdiv
udisp_b,vdisp_b = uvor, vvor
udisp_c,vdisp_c = udef1, vdef1
udisp_d,vdisp_d = udef2, vdef2

# Display the vector field as arrows, CENTERED ON the point they apply to
fig = plt.subplots(figsize=(8,8))

ax1 = plt.subplot(2,2,1)
ax1.quiver(x,y,udisp_a,vdisp_a,pivot='mid')
ax1.set_title('Pure unit divergence')

ax1 = plt.subplot(2,2,2)
ax1.quiver(x,y,udisp_b,vdisp_b,pivot='mid')
ax1.set_title('Pure unit vorticity')

ax1 = plt.subplot(2,2,3)
ax1.quiver(x,y,udisp_c,vdisp_c,pivot='mid')
ax1.set_title('Pure unit deformation 1')

ax1 = plt.subplot(2,2,4)
ax1.quiver(x,y,udisp_d,vdisp_d,pivot='mid')
ax1.set_title('Pure unit deformation 2')

plt.tight_layout()
plt.show()
```
